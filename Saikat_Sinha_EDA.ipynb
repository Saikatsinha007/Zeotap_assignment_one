```python
# Ensure TransactionDate and SignupDate are in datetime format
merged_df["TransactionDate"] = pd.to_datetime(merged_df["TransactionDate"])
merged_df["SignupDate"] = pd.to_datetime(merged_df["SignupDate"])

# Calculate the first transaction date for each customer
customer_first_transaction = merged_df.groupby("CustomerID")["TransactionDate"].min().reset_index()
customer_first_transaction.rename(columns={"TransactionDate": "FirstTransactionDate"}, inplace=True)

# Calculate total transaction count per customer
customer_transaction_count = merged_df.groupby("CustomerID")["TransactionID"].count().reset_index()
customer_transaction_count.rename(columns={"TransactionID": "TransactionCount"}, inplace=True)

# Merge these metrics back into the original dataset
customer_analysis = pd.merge(customer_first_transaction, customer_transaction_count, on="CustomerID")
customer_analysis = pd.merge(customer_analysis, merged_df[["CustomerID", "SignupDate"]].drop_duplicates(), on="CustomerID")

# Calculate time difference between signup and first transaction
customer_analysis["DaysToFirstTransaction"] = (customer_analysis["FirstTransactionDate"] - customer_analysis["SignupDate"]).dt.days
```

```python
customer_analysis["DaysToFirstTransaction"].plot(kind="hist", bins=30, color="skyblue", edgecolor="black", title="Days to First Transaction")
plt.xlabel("Days to First Transaction")
plt.ylabel("Frequency")
plt.show()
```

### Interpretation of Days to First Transaction Histogram

1. **Early Engagement**: Many customers make their first transaction within the first 100 days, highlighting the importance of early engagement.
2. **Gradual Decline**: As time goes on, fewer customers make their first transaction, indicating a drop in conversion.
3. **Long-Tail Behavior**: Some customers take up to 1000 days to make a purchase, suggesting the need for targeted outreach.

### Insights

1. **Marketing Strategies**: Focus on **early engagement** with targeted offers and incentives to boost first purchases within the first 100 days.
2. **Customer Segmentation**: Address **long-tail customers** with personalized campaigns, as they may need different triggers to convert.
3. **Customer Retention**: Compare retention between early and late purchasers to tailor retention strategies and nurture long-term loyalty.

```python
# Dynamically set bins and labels
bins = [0, 5, 10, 15]
labels = ["1-5", "6-10", "11+"]
customer_analysis["TransactionCountCategory"] = pd.cut(
    customer_analysis["TransactionCount"], bins=bins, labels=labels, right=False
)

# Calculate average DaysToFirstTransaction for each category
avg_days = customer_analysis.groupby("TransactionCountCategory")["DaysToFirstTransaction"].mean()

# Plot bar chart
avg_days.plot(kind="bar", color="skyblue", edgecolor="black", title="Average Days to First Transaction by Transaction Count Category")
plt.xlabel("Transaction Count Category")
plt.ylabel("Average Days to First Transaction")
plt.xticks(rotation=0)
plt.show()
```

```python
# Group data by Region and Category and calculate counts
region_category = merged_df.groupby(["Region", "Category"]).size().unstack(fill_value=0)

# Calculate percentages
region_category_percentage = region_category.div(region_category.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = region_category.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="Set2")
plt.title("Region-wise Category Distribution (with Percentages)")
plt.xlabel("Region")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=45)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")

# Annotate percentages on each block of the bar
for i, region in enumerate(region_category.index):
    bottom = 0
    for category in region_category.columns:
        value = region_category.loc[region, category]
        percentage = region_category_percentage.loc[region, category]
        if value > 0:  # Only annotate non-zero values
            ax.text(i, bottom + value / 2, f"{percentage:.1f}%", ha="center", va="center", fontsize=9)
        bottom += value

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to format numbers (e.g., 5000 -> 5k)
def human_readable(num):
    if num >= 1_000_000:
        return f'{num / 1_000_000:.1f}M'
    elif num >= 1_000:
        return f'{num / 1_000:.1f}k'
    return str(num)

# Group by Region and Category, then calculate the total sales (TotalValue)
region_category_sales = merged_df.groupby(["Region", "Category"])["TotalValue"].sum().unstack()

# Plotting the bar plot
ax = region_category_sales.plot(kind="bar", stacked=True, figsize=(10, 6))

# Adding numbers inside each bar with percentages
for i, p in enumerate(ax.patches):
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2  # X position to center the number
    y = p.get_y() + height / 2  # Y position to center the number
    region = region_category_sales.index[i // len(region_category_sales.columns)]
    total_sales = region_category_sales.loc[region].sum()  # Total sales for the region
    percentage = (height / total_sales) * 100  # Calculate percentage
    
    # Add percentage and raw value as text inside the bar
    ax.text(x, y, f'{percentage:.1f}%\n{human_readable(height)}', ha='center', va='center', fontsize=10, color='white')

print("\nTotal Sales by Region and Category")
print(region_category_sales)
# Adding title and labels
plt.title("Total Sales by Region and Category")
plt.xlabel("Region")
plt.ylabel("Total Sales (USD)")

# Adding legend
plt.legend(title="Category")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to format numbers (e.g., 5000 -> 5k)
def human_readable(num):
    if num >= 1_000_000:
        return f'{num / 1_000_000:.1f}M'
    elif num >= 1_000:
        return f'{num / 1_000:.1f}k'
    return str(num)

# Category-wise Quantity
category_quantity = merged_df.groupby('Category')['Quantity'].sum().reset_index()

# Category-wise TotalValue
category_total_value = merged_df.groupby('Category')['TotalValue'].sum().reset_index()

# Set up the figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot Category-wise Quantity
sns.barplot(x='Category', y='Quantity', data=category_quantity, ax=axes[0], palette='viridis')
axes[0].set_title("Category-wise Total Quantity")
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Total Quantity")
for p in axes[0].patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2  # X position to center the number
    y = p.get_y() + height / 2  # Y position to center the number
    axes[0].text(x, y, f'{human_readable(height)}', ha='center', va='center', fontsize=10, color='white')

# Plot Category-wise TotalValue
sns.barplot(x='Category', y='TotalValue', data=category_total_value, ax=axes[1], palette='magma')
axes[1].set_title("Category-wise Total Value")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Total Value")
for p in axes[1].patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2  # X position to center the number
    y = p.get_y() + height / 2  # Y position to center the number
    axes[1].text(x, y, f'{human_readable(height)}', ha='center', va='center', fontsize=10, color='white')

plt.tight_layout()
plt.show()
```
